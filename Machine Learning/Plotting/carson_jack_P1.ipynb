### No xrange

The xrange from Python 2 is now merged into "range" for Python 3 and there is no xrange in Python 3. In Python 3, range(3) does not create a list of 3 elements as it would in Python 2, rather just creates a more memory efficient iterator.

Hence,
xrange in Python 3: Does not exist
range in Python 3: Has very similar behavior to Python 2's xrange

````python
for i in range(3):
    print(i)
````

````python
##############################################################################
#          TODO: Use the loop function for plotting figures                 #
##############################################################################
import matplotlib.pyplot as plt

def get_feature_codes(): # Get input from user
    x = int(input("Enter feature code For the horizontal axis: "))
    y = int(input("Enter feature code For the vertical axis: "))
    return x, y

def plot_feature_map(): # Check if input is valid and return feature codes
    valid = [0,1,2,3]
    print("You can do a plot of any two features of the Iris Data set The feature codes are:\n0 = sepal length\n1 = sepal width\n2 = petal length\n3 = petal width\n")    
    while True:
        x, y = get_feature_codes()
        if x in valid and y in valid:
            break
        else:
            print("Please enter valid feature codes")
    return x, y 

def show_plot(): # Plot the feature code data
    plt.title("Iris Flower Plot")
    x, y = plot_feature_map() 
    plt.xlabel(labels[x])
    plt.ylabel(labels[y])
    for i in range(len(setosa_data)):
        plt.scatter(setosa_data[i][x], setosa_data[i][y], color = "green", marker = 'v', label = "Setosa" if i==0 else "")
    for i in range(len(versicolor_data)):
        plt.scatter(versicolor_data[i][x], versicolor_data[i][y], color = "blue", label = "Versicolor" if i==0 else "")
    for i in range(len(virginica_data)):
        plt.scatter(virginica_data[i][x], virginica_data[i][y], color = "red", marker = '+', label = "Virginica" if i==0 else "")
    plt.legend()
    plt.show()

data = []
setosa_data = []
virginica_data = []
versicolor_data = []
labels = ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]

with open("IrisData.txt", 'r') as f: # Read data from file
    for line in f:
        line_data = line.split('\t')
        data.append(line_data)

data.pop(0) # Remove the first line of the file
for i in range(len(data)): # Strip \n and separate Iris Types
    data[i][-1] = data[i][-1].strip()
    if data[i][-1] == "setosa":
        setosa_data.append(data[i])
    elif data[i][-1] == "virginica":
        virginica_data.append(data[i])
    elif data[i][-1] == "versicolor":
        versicolor_data.append(data[i])
    
for i in range(len(data)): # Convert to float
    for j in range(4):
        data[i][j] = float(data[i][j])

while True: # Plot loop 
    show_plot()
    inp = input("\nWould you like to do another plot? (y/n) ")
    if inp == 'y':
        continue
    else:
        break
##############################################################################
#                              END OF YOUR CODE                              #
##############################################################################
````
